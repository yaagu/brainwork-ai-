```python
# создание массивов для поэлементарного сложения и умножения
arrq = np.array([
    [7, 8, 9],
    [3, 4, 2],
    [4, 9, 3]
])

arrw = np.array([
    [3, 7, 1],
    [6, 9, 4],
    [3, 1, 8]
])
```

```python
# сложение
arrq + arrw
```

```python
# умножение
arrq * arrw
```

```python
# поиск среднего
arr_sum_mean = (arrq + arrw).mean()
arr_sum_mean.item()
```
